```python
def line_to_tensor(word):
  # returns N by 1 by 300
  words  = word.split() 
  sub_matrix = np.array(());
  max_len = min(20, len(words))
  # print("max_len = " + str(max_len))
  for i in range(0, max_len):
    # print(sub_matrix.shape)

    word = words[i];
    # print(word)
    if word in wv.vocab: 
      # print("vector shape=" + str(wv.get_vector(word).shape))
      if i == 0: 
        sub_matrix = wv.get_vector(word).copy()
        # print(sub_matrix.shape)
      else:
        # print("sub shape =" + str(sub_matrix.shape))
        sub_matrix = np.vstack((sub_matrix,  wv.get_vector(word)))
        # print("sub shape =" + str(sub_matrix.shape))

    else: 
      # padd with zero 
      # np.concatenate((sub_matrix,  np.zeros(300)), axis=None)
      if i == 0: 
        sub_matrix = np.zeros(300)
      else:
        sub_matrix = np.vstack((sub_matrix,  np.zeros(300)))
    
  for pad in range(max_len, 20):
    # padd with zero 
    if pad == 0: 
        sub_matrix = np.zeros(300)
    else:
      sub_matrix = np.vstack((sub_matrix,  np.zeros(300)))

  if (sub_matrix.shape[0]!= 20  and sub_matrix.shape[1]!= 300):
    print("sub shape =" + str(np.size(sub_matrix.shape)))
    print("size of sequence is not 300 * 20")
    return
  t = torch.tensor(sub_matrix)
  # print("created tensor of shape = "+str(t.shape))
  return t
```

```python
input_size = 300
hidden_size = 20
num_layers = 2
num_classes = 2

import torch.nn as nn


class BiGRU(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers, num_classes):
        super(BiGRU, self).__init__()
        self.hidden_size = hidden_size
        self.num_layers = num_layers
        self.gru = nn.GRU(input_size, hidden_size, num_layers, batch_first=True, bidirectional=True)
        self.fc = nn.Linear(hidden_size*2, num_classes)  # 2 for bidirection
    
    def forward(self, x):
        # Set initial states
        # print(x)
        h0 = torch.zeros(self.num_layers*2, x.size(0), self.hidden_size).to(device) # 2 for bidirection 
        # print(h0.shape)

        c0 = torch.zeros(self.num_layers*2, x.size(0), self.hidden_size).to(device)
        # print(c0.shape)

        # Forward propagate GRU
        out, hn = self.gru(x, h0)  # out: tensor of shape (batch_size, seq_length, hidden_size*2)
        # print(out.shape)
        # Decode the hidden state of the last time step
        out = self.fc(out[:, -1, :])
        return out

model = BiGRU(input_size, hidden_size, num_layers, num_classes).to(device)
# loss_fn = nn.CrossEntropyLoss()
```

```python
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.SGD(model.parameters(), lr=1e-3)
num_epochs = 150
n_total_steps = len(dl_train_)
# b_size= find_largest_divisor(dl_train_)
# print(b_size)

for epoch   in range(num_epochs):
    for i, (reviews, labels) in enumerate(dl_train_):
      # print(reviews)
      reviews_t = getBatchTensor(reviews, bsize_train, 20, 300).to(device)
      # print(labels)
      labels_t = labels.to(device)

      # Forward pass
      outputs = model(reviews_t.float())
      loss = criterion(outputs, labels_t)
      # print("loss"+str(loss))
      # Backward and optimize
      optimizer.zero_grad()
      loss.backward()
      optimizer.step()


      if (i+1) % 2000 == 0:
          print (f'Epoch [{epoch+1}/{num_epochs}], Step [{i+1}/{n_total_steps}], Loss: {loss.item():.4f}')        
```
